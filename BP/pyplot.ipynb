```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#p=0.01
datap0p005_0p015= pd.read_csv('p0.005-0.015_data.txt',sep='\s+',header=None)
datap0p01_0p01=pd.read_csv('p0.01-0.01_data.txt',sep='\s+',header=None)
datap0p008_0p012=pd.read_csv('p0.008-0.012_data.txt',sep='\s+',header=None)

x=datap0p005_0p015[0]
y=datap0p005_0p015[1]
x2=datap0p01_0p01[0]
y2=datap0p01_0p01[1]
x3=datap0p008_0p012[0]
y3=datap0p008_0p012[1]

plt.xlabel('n')
plt.ylabel('failure rate')

plt.plot(x, y,"r",label="avg=0.01, range=0.005")#0.005-0.015  
plt.plot(x2, y2,"b",label="0")
plt.plot(x3, y3,"y",label="avg=0.01,range=0.002")#0.008-0.012
plt.legend()
plt.show()
#use #/frac of fails
```

```python
#n=128
datan128= pd.read_csv('n128_data.txt',sep='\s+',header=None)
datan128sp= pd.read_csv('n128sp_data.txt',sep='\s+',header=None)
x=datan128[2]
y=datan128[1]
x2=datan128sp[2]
y2=datan128sp[1]
plt.xlabel('p')
plt.ylabel('# of sec decoding')
plt.plot(x, y,"r",label="diff p:(0.004i,0.006i)")
plt.plot(x2, y2,"b",label="same p:(0.005i,0.005i)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot2(data_file1,data_file2,x_index,y_index,xlabel,ylabel,data1_label,data2_label,logx,logy):
    data1= pd.read_csv(data_file1,sep='\s+',header=None)
    data2=pd.read_csv(data_file2,sep='\s+',header=None)
    
    x1=data1[x_index]
    y1=data1[y_index]
    x2=data2[x_index]
    y2=data2[y_index]
    
    if logx!=0:
        x1=np.log(x1)
        x2=np.log(x2)
    if logy!=0:
        y1=np.log(y1)
        y2=np.log(y2)
        
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.plot(x1, y1,"r",label=data1_label)
    plt.plot(x2, y2,"b",label=data2_label)
    
    plt.legend()
    plt.show()
  
```

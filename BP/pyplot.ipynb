```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#p=0.01
datap0p01= pd.read_csv('p0.01_data.txt',sep='\s+',header=None)
datap0p01sp=pd.read_csv('p0.01sp_data.txt',sep='\s+',header=None)
x=datap0p01[0]
y=datap0p01[1]
x2=datap0p01sp[0]
y2=datap0p01sp[1]
plt.xlabel('n')
plt.ylabel('# of sec decoding')
plt.plot(x, y,"r",label="diff p")
plt.plot(x2, y2,"b",label="same p")
plt.legend()
plt.show()
```

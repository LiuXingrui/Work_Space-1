```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
   
def logplotn(n,data_file_list,x_index,y_index,xlabel,ylabel,data_label_list,title,lfit=0):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title) 
    plt.grid()
    
    for i in range(n):
        data=pd.read_csv(data_file_list[i],sep='\s+',header=None)
        x=data[x_index]
        y=data[y_index]
        x=np.log(x)
        y=np.log(y)
        plt.plot(x, y,label=data_label_list[i])
    
    if lfit!=0:
        fit=lfit*x
        plt.plot(x, fit,label="linear fit: slope=%f"%lfit)
        
    plt.legend()
    plt.show()

def plotn(n,data_file_list,x_index,y_index,xlabel,ylabel,data_label_list,title,lfit=0):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title) 
    plt.grid()
    
    for i in range(n):
        data=pd.read_csv(data_file_list[i],sep='\s+',header=None)
        x=data[x_index]
        y=data[y_index]
        plt.plot(x, y,label=data_label_list[i])
        
    if lfit!=0:
        fit=lfit*x
        plt.plot(x, fit,label="linear fit: slope=%f"%lfit)
        
    plt.legend()
    plt.show()
    

  
```

```python
#p=0.01, #index: n=0, fail_rate=1, wt/avg_p=2, avg_iter=3 number_of_suc_decoding=4
n=3
data_file1='p0.005-0.015_data.txt'
data_file2='p0.01-0.01_data.txt'
data_file3='p0.008-0.012_data.txt'
data_file_list=[data_file1,data_file2,data_file3]

xlabel='n'
ylabel='fail rate'

data1_label="avg=0.01, range=0.005"
data2_label="0"
data3_label="avg=0.01,range=0.002"
data_label_list=[data1_label,data2_label,data3_label]

x_index=0
y_index=1
title="diff p vs same p"

plotn(n,data_file_list,x_index,y_index,xlabel,ylabel,data_label_list,title)

#use #/frac of fails
```
